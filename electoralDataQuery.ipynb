```python
import pandas as pd

# Load the CSV file
file_path = 'combined.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
def get_total_amount_purchased_by_name_and_party(name):
    # Filter the dataframe by the given name
    filtered_df = df[df['purchaser_name'] == name]

    # Group by political party and calculate the total amount for each party
    total_amount_by_party = filtered_df.groupby('political_party')['denomination'].sum().reset_index()

    # Convert the amount to lakhs
    total_amount_by_party['denomination'] = total_amount_by_party['denomination'] / 100000

    # Sort the results in descending order
    total_amount_by_party = total_amount_by_party.sort_values(by='denomination', ascending=False)

    return total_amount_by_party

# Example usage
name = 'SHIRDI SAI ELECTRICALS LTD'
total_amount_by_party = get_total_amount_purchased_by_name_and_party(name)
print(f"Total amount purchased by {name} for each party (in lakhs):\n{total_amount_by_party}")
```
